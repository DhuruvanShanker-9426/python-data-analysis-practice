```python
languages = ['C','C++','Java','Python','R']
students = [200,250,350,400,150]
colors = ['green','red','purple','blue','orange']
```

```python
plt.figure(figsize=(7,5),facecolor='lightgrey')
plt.bar(languages,students,color=colors)
plt.title('Number of Students in different Languages')
plt.xlabel('Languages')
plt.ylabel('Number of Students')
ax = plt.gca()
ax.set_facecolor('lightblue')
```

```python
plt.figure(figsize=(7,5),facecolor='lightgrey')
plt.barh(languages,students,color=colors)
plt.title('Number of Students in different Languages')
plt.xlabel('Number of Students')
plt.ylabel('Languages')
ax = plt.gca()
ax.set_facecolor('lightblue')
```

```python
plt.pie(students,labels=languages,shadow=True,autopct='%1.2f%%',explode=(0,0,0,0.1,0),startangle=100)
plt.show()
```

```python
x = [20, 22, 25, 27, 30, 32, 35]
y = [200, 220, 260, 280, 310, 330, 360]
category = ['City A', 'City A', 'City B', 'City B', 'City C', 'City C', 'City C']
value = [1, 2, 3, 4, 5, 6, 7]

# Create scatter plot
sns.scatterplot(
    x=x,
    y=y,
    hue=category,   # color based on category
    size=value,     # size based on value
    sizes=(40,200),# control min and max dot size
    alpha=0.5
)

# Labels and title
plt.title("Temperature vs Ice Cream Sales")
plt.xlabel("Temperature (°C)")
plt.ylabel("Ice Cream Sales")

plt.show()
```

```python
data = [10, 20, 30, 40, 50,100]
plt.boxplot(data,label="Datas",patch_artist=True,
            boxprops=dict(facecolor='lightgrey',color='red'),
    medianprops=dict(color='yellow'),
    whiskerprops=dict(color='green'))
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6, 3), subplot_kw=dict(aspect="equal"))

recipe = ["375 g flour",
          "75 g sugar",
          "250 g butter",
          "300 g berries"]

data = [float(x.split()[0]) for x in recipe]
ingredients = [x.split()[-1] for x in recipe]


def func(pct, allvals):
    absolute = int(np.round(pct/100.*np.sum(allvals)))
    return f"{pct:.1f}%\n({absolute:d} g)"


wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"))

ax.legend(wedges, ingredients,
          title="Ingredients",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=8, weight="bold")

ax.set_title("Matplotlib bakery: A pie")

plt.show()
```
